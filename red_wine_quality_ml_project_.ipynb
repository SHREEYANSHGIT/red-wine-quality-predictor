importing the  dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

data collection


```python
#loading the dataset to a pandas data frames
wine_dataset = pd.read_csv('winequality-red.csv')
```

```python
#numberof rows and columns
wine_dataset.shape
```

```python
# first 5 rows of dataset
wine_dataset.head()
```

```python
#checking missing values for data set
wine_dataset.isnull().sum()
```

data analysis and visulaization

```python
#statistical measures of data set
wine_dataset.describe()
```

```python
#number of value for each quality
sns.catplot(x='quality',data=wine_dataset,kind='count')
```

```python
# volatile acidity vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality',y='volatile acidity',data=wine_dataset)
```

```python
# citric acid vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality',y='citric acid',data=wine_dataset)
```

Correlation
1. positive correlation # directly prop
2. negative correlation # inversely prop

```python
correlation = wine_dataset.corr()
```

```python
# constructing heatmap to under stand correlation b/w the columns
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Reds')
```

Data preprocessing


```python
#seperation of data
X = wine_dataset.drop('quality',axis=1)
```

```python
print(X)
```

label binarization


```python
Y = wine_dataset['quality'].apply(lambda Y_value : 1 if Y_value>=7 else 0)
```

```python
print(Y)
```

data spliting in training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model training


```python
#random forest classifier
model = RandomForestClassifier(n_estimators=100)

```

```python
model.fit(X_train,Y_train)
```

model evaluation
accuarcy Score


```python
#accuarcy  on test data
X_test_prediction = model.predict(X_test)
test_data_accuarcy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuarcy :',test_data_accuarcy)
print('Accuarcy :',test_data_accuarcy)
```

```python
import pickle

# suppose 'model' is your trained model
with open("model.pkl", "wb") as file:
    pickle.dump(model, file)
```
